3) Queremos hacer un traductor que cambie las palabras masculinas de una frase
por su versión neutra. Como primera aproximación, completá el siguiente código
para reemplazar todas las letras 'o' que figuren en el último o anteúltimo caracter
de cada palabra por una 'e'.
Por ejemplo 'todos somos programadores' pasaría a ser 'todes somes
programadores'.



```python
def neutralizar_palabras(cadena):
  palabras = cadena.split() #Acá paso de cadena de caracteres a lista
  for i in range(len(palabras)): #Pasé a lista porque sino todo esto no lo puedo hacer
    if len(palabras[i]) > 2 and palabras[i][-2] == 'o':
      palabras[i] = palabras[i][:-2] + 'e' + palabras[i][-1]
    elif len(palabras[i]) > 2 and palabras[i][-1] == 'o':
      palabras[i] = palabras[i][:-1] + 'e'
  return ' '.join(palabras) #Acá paso de lista a cadena de caracteres

frase = 'Somos tan distintos e iguales'
print(neutralizar_palabras(frase))
```

8) Definir una función cant_e que tome una lista de caracteres y devuelva la
cantidad de letras ‘e’ que tiene la misma.


```python
def cantidad_de_e(frase):
    res = 0
    palabras = frase.split()
    for i in range(len(palabras)):
        for j in range(len(palabras[i])):
            if palabras[i][j] == 'e':
                res+=1
    return res

frase = 'Somos tan distintos e iguales'
print(cantidad_de_e(frase))

neutralizar = neutralizar_palabras(frase)
print(cantidad_de_e(neutralizar))
```

Definir la función mezclar(cadena1, cadena2) que tome dos strings y
devuelva el resultado de intercalar elemento a elemento. Por ejemplo: si
intercalamos Pepe con Jose daría PJeopsee. En el caso de Pepe con Josefa daría
PJeopseefa.


```python
def mergear_nombres(nombre1, nombre2):
    juntos = ''

    if len(nombre1) <= len(nombre2):
        for i in range(len(nombre1)):
            juntos += nombre1[i]
            juntos += nombre2[i]
        for j in range(len(nombre1),len(nombre2)):
            juntos += nombre2[j]
    else:
        for i in range(len(nombre2)):
            juntos += nombre1[i]
            juntos += nombre2[i]
        for j in range(len(nombre2),len(nombre1)):
            juntos += nombre1[j]
    return juntos

string1 = 'Agustin'
string2 = 'Dafne'
print(mergear_nombres(string1, string2) + ' Te amo hermosote')
```

Construí una función traductor_geringoso(lista) que, a partir de una lista de palabras, devuelva un
diccionario geringoso.
Las claves del diccionario deben ser las palabras de la lista y los valores deben ser sus traducciones al geringoso.

```python
def traductor_geringoso(lista):
    traducciones = {}
    for i in range(len(lista)):
        traducciones[lista[i]] = geringoso(lista[i])
    return traducciones

def geringoso(palabra):
  capadepenapa = ''
  palabra.split(',')
  vocales = {'a', 'e', 'i', 'o', 'u'}
  for c in palabra:
    capadepenapa += c
    if c in vocales:
      capadepenapa += 'p' + c
  return capadepenapa

lista = ['banana', 'manzana', 'mandarina']
print(traductor_geringoso(lista))
```

Definir una función pisar_elemento(M,e) que tome una matriz de enteros M y un entero e y devuelva
una matriz similar a M donde las entradas coincidentes con e fueron cambiadas por -1.
Por ejemplo si M = np.array([[0, 1, 2, 3], [4, 5, 6, 7]]) y e = 2, entonces la función debe devolver la matriz
np.array([[0, 1, -1, 3], [4, 5, 6, 7]])

```python
import numpy as np
import copy

def pisar_elemento(M,e):
    copia = copy.deepcopy(M)
    for i in range(len(copia)):
        for j in range(len(copia[0])):
            if(copia[i][j] == e):
                copia[i][j] = -1
    return copia


M = np.array([[1,2,3],[1,2,3]])
e = 2
print(pisar_elemento(M, e))
```
